```python
import os
import yaml
import json

# Function to check for the 'addressMatch' in the frontmatter
def get_address_match(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        # Read the frontmatter section
        content = f.read()
        if content.startswith('---'):
            frontmatter_end = content.find('---', 3)
            frontmatter = content[3:frontmatter_end]
            frontmatter_data = yaml.safe_load(frontmatter)
            # Check if 'addressMatch' exists in the frontmatter
            if 'addressMatch' in frontmatter_data:
                return frontmatter_data['addressMatch']
    return None

# List to hold the dictionaries for files with 'addressMatch'
address_match_list = []

# Walk through the directory tree
for root, dirs, files in os.walk('docs'):
    for file in files:
        if file.endswith('.md'):
            file_path = os.path.join(root, file)
            address_match = get_address_match(file_path)
            if address_match:
                # Create the dict and append it to the list
                address_match_list.append(
                    {
                        "addressMatch": address_match,
                        "file": file_path
                    }
                )

# # Print the final list of dictionaries
# for item in address_match_list:
#     # print(item)
#     json_output = json.dumps(item, indent=2)
#     print(json_output)

address_match_list = sorted(address_match_list, reverse=True, key=lambda x: len(x['addressMatch']))

print(json.dumps(address_match_list, indent=2) )
```

```python
import os
import yaml
import shutil
import json


# Function to check for the 'addressMatch' in the frontmatter
def get_address_match(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        # Read the frontmatter section
        content = f.read()
        if content.startswith('---'):
            frontmatter_end = content.find('---', 3)
            frontmatter = content[3:frontmatter_end]
            frontmatter_data = yaml.safe_load(frontmatter)
            # Check if 'addressMatch' exists in the frontmatter
            if 'addressMatch' in frontmatter_data:
                return frontmatter_data['addressMatch']
    return None


# List to hold the dictionaries for files with 'addressMatch'
address_match_list = []

# The source directory
source_dir = 'docs'

# The destination directory 
dest_dir = 'embedded_docs/docs'  

# Walk through the directory tree and copy files from matching directories
for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.endswith('.md'):
            file_path = os.path.join(root, file)
            address_match = get_address_match(file_path)
            if address_match:
                # Get the relative path of the directory containing the matching .md file
                relative_dir_path = os.path.relpath(root, source_dir)
                dest_dir_path = os.path.join(dest_dir, relative_dir_path)

                # Create the dict and append it to the list
                address_match_list.append(
                    {
                        "addressMatch": address_match,
                        "file": file_path
                    }
                )

                # Ensure the destination directory exists
                os.makedirs(dest_dir_path, exist_ok=True)

                # Copy all files from the directory where the matching .md file is located
                for dir_file in os.listdir(root):
                    source_file_path = os.path.join(root, dir_file)
                    dest_file_path = os.path.join(dest_dir_path, dir_file)

                    # If it's a file, copy it
                    if os.path.isfile(source_file_path):
                        shutil.copy2(source_file_path, dest_file_path)
                        print(f"Copied: {source_file_path} -> {dest_file_path}")
                    # If it's a directory, copy the entire directory
                    elif os.path.isdir(source_file_path):
                        shutil.copytree(source_file_path, dest_file_path, dirs_exist_ok=True)
                        print(f"Copied directory: {source_file_path} -> {dest_file_path}")
```

```python
# # Print the final list of dictionaries
# for item in address_match_list:
#     # print(item)
#     json_output = json.dumps(item, indent=2)
#     print(json_output)
address_match_list = sorted(address_match_list, reverse=True, key=lambda x: len(x['addressMatch']))
print(json.dumps(address_match_list, indent=2) )
```
